```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# import data from output file as csv
timeMatrix = pd.read_csv("output.txt", header=None, delimiter=" ")
timeMatrix=timeMatrix.T
timeMatrix
```

```python
c = ['r--o', 'g--o', 'b--o', 'y--o']
plt.subplots(figsize=(10, 7))
plt.suptitle('Time vs Iteration')
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.plot(range(1, len(timeMatrix) + 1), timeMatrix.iloc[:, i], c[i])
    plt.title(timeMatrix.columns[i])
    plt.grid()
    plt.xticks(range(1, 4, 1))
    plt.xlabel('Iteration')
    plt.ylabel('Time (s)')
plt.tight_layout()
```
