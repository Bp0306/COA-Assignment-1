```python
plt.figure(figsize=(7, 4))
plt.plot(n_values, py_INT_system, 'r-o', label='Python INT system time', markersize=8, linewidth=2)
plt.plot(n_values, cpp_INT_system, 'b-o', label='C++ INT system time', markersize=8, linewidth=2)

plt.xlabel('Matrix Size (N)', fontsize=12)
plt.ylabel('System Time (s)', fontsize=12)
plt.xticks(n_values, fontsize=10)
plt.yticks(fontsize=10)

plt.title('System Time Comparison: Python vs C++ (Integer)', fontsize=14, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```
